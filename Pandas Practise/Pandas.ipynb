## 13) Filter data:

<br>

**1) Which cars have highest mpg?(display model_name and mpg of it)**

```python
top_mpg = car.sort_values('mpg',ascending = False)[0:5]
```

```python
top_mpg[['model_name','mpg']]
```

**2) How many cars have 4 gears?**

```python
car.loc[car['gear']==4,'gear'].count()
```

```python
car.loc[car['gear']==4,'gear'].value_counts()
```

**3) What is the highest displacement mentioned in a dataset?(print model_name and its displacment)**

```python
max_disp = car['disp'].max()
car.loc[car['disp'] == max_disp,['model_name','disp']]
```
